```python
#Loading the dataset and import the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
 

df = pd.read_csv("/Users/berkeela/Desktop/house/archive-2/Melbourne_housing_FULL.csv")
print(df.shape)
print(df.info())
```

```python
#Getitng the datasets memory usage and take a look at it
print(df.memory_usage())
df.head(20)
```

```python
#Drop the duplicated data and fill the empty cell with mode
df.drop_duplicates(inplace = True)

cols = df.columns

num_cols = df._get_numeric_data().columns

categorical_columns = list(set(cols) - set(num_cols))

for column in cols:
    df[column] = df[column].fillna(df[column].mode().iloc[0])

print(categorical_columns)

```

```python
from scipy.stats import zscore
#Getting rid of the outliers in the Landsize and BuildingArea columns of the dataset
threshold = 4
clean_data = df


z_scores = zscore(clean_data["Landsize"])

outlier_indices = np.where(z_scores > threshold)

clean_data = clean_data[z_scores <= threshold]



z_scores = np.abs(zscore(clean_data["BuildingArea"]))

outlier_indices = np.where(z_scores > threshold)

clean_data = clean_data[z_scores <= threshold]




print("Indices of outliers:", outlier_indices)

print("Cleaned data:", clean_data)



```

```python
#Implementation of the label encoding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

non_numerical_columns = ["Suburb","Address","Rooms","Type","Method","SellerG","CouncilArea","Regionname"]

label_encoded_data = clean_data

for column in non_numerical_columns:
    encoded_data = label_encoder.fit_transform(label_encoded_data[column])
    label_encoded_data[column] = encoded_data

label_encoded_data.head(30)
```

```python
#Implementation of the one-hot encoding
from sklearn.preprocessing import OneHotEncoder
one_hot_encoder = OneHotEncoder(sparse=False)
one_hot_encoded_data = clean_data

for column in non_numerical_columns:
    encoded_data = one_hot_encoder.fit_transform(np.array(one_hot_encoded_data[column]).reshape(-1, 1))
    one_hot_encoded_data[column] = encoded_data

one_hot_encoded_data.head(50)
```

```python
#As you can see the one-hot encoding is not suitable at least not proper for our dataset so, we're gonna go with the label encoding 
#We're gonna do some price prediction so our x = price column(that also means the price column of our dataset is our target)
#Here's the test train seperation and decleration of the target column
#Dtae column could be a pain in the neck so it dropped out 
from sklearn.model_selection import train_test_split

label_encoded_data = label_encoded_data.drop("Date", axis=1)
X = label_encoded_data.drop("Price", axis=1) 
y = label_encoded_data["Price"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor

models = { 
'Lasso': {
'model': Lasso() },

'LinearRegression': {
'model': LinearRegression()
},

'Ridge': {
'model': Ridge() },

'ElasticNet': {
'model': ElasticNet()
},

'KNeighborsRegressor': {
'model': KNeighborsRegressor() },

'RandomForestRegressor': {
'model': RandomForestRegressor()
},
'GradientBoostingRegressor': {
'model': GradientBoostingRegressor() },

'AdaBoostRegressor': {
'model': AdaBoostRegressor(n_estimators = 5, learning_rate = 1.2, loss = 'exponential', random_state = 2)
}

}

results = []

for regressors in models.values():
    for model in regressors.values():

        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        results.append(y_pred)

print(results)
```

```python
#Model Evaluation
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

regressor_names = ["Lasso","LinearRegression","Ridge","ElasticNet","KNeighborsRegressor","RandomForestRegressor","GradientBoostingRegressor","AdaBoostRegressor"]
regresor_name_iterator = 0

for y_pred in results:

    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)


    print(regressor_names[regresor_name_iterator])
    print("mean_squared_error: "+ str(mse))
    print("mean_absolute_error: "+ str(mae))
    print("root_mean_squared_error:"+ str(mae))
    print("r2"+ str(r2))
    print("\n")
    regresor_name_iterator = regresor_name_iterator + 1



```

With these number it's clear to see RandomForestRegressor has the lowest error rate but the GradientBoostingRegressor also has noticeable statistics for this dataset.
